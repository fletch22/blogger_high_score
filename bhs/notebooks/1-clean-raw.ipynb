```python
from bhs.config import constants

file_path = constants.ANALYST_STOCK_PICKS_FROM_TICKER_PATH

import pandas as pd
df = pd.read_parquet(file_path)

print(df.shape[0])    
```

```python
from bhs.utils import date_utils

df_rat = df

def convert_dt(date_str):
    dt = date_utils.parse_tipranks_dt(date_str=date_str)
    return date_utils.get_standard_ymd_format(date=dt)
    
    
df_rat["dt_rating"] = df_rat.apply(lambda x: convert_dt(x["dt_rating"]), axis=1)


df_rat.head(100)
```

```python
# parse rank_raw and / 5

df_ranked = df_rat

def convert_rank(rank_raw):
    return float(rank_raw.split(" ")[0])

df_ranked["rank_raw"] = df_ranked["rank_raw"].fillna("2.5 out of 5")
df_ranked["rank"] = df_ranked.apply(lambda x: convert_rank(x["rank_raw"]), axis=1)
df_ranked = df_ranked.drop(columns=["rank_raw"])

rank_unique_cols = ["ticker", "analyst_name", "dt_rating"]
df_ranked = df_ranked.sort_values(by=rank_unique_cols).drop(columns="analyst_rel_url")

def convert_rating(rating_str):
    result = 0
    if rating_str == "buy":
        result = 1
    elif rating_str == "sell":
        result = -1
    return result

df_ranked["rating_raw"] = df_ranked["rating"]
df_ranked["rating"] = df_ranked["rating_raw"].apply(convert_rating)

df_ranked["target_price"] = df_ranked["target_price"].str.replace('$', '').replace('None', None)

df_ranked.head(5)
```

```python
from bhs.services import ticker_service, tip_ranks_service
from datetime import timedelta
# Calc rating_roi on date of rating

ranked_tickers = df_ranked["ticker"].unique()

all_tick = []
for rt in ranked_tickers:
    df = ticker_service.get_ticker_eod_data(rt)
    
    df_rt = df_ranked[df_ranked["ticker"] == rt]
   
    min_date = df_rt["dt_rating"].min()
    max_date = df_rt["dt_rating"].max()
    
    dt_max = date_utils.parse_std_datestring(max_date)
    dt_max = dt_max + timedelta(days=5)
    max_date = date_utils.get_standard_ymd_format(dt_max)
    
    df = df[(df["date"] > min_date) & (df["date"] < max_date)]
    df["next_day"] = df["date"]
    df[["next_day"]] = df[["next_day"]].shift(-1)
    
    df_merged = pd.merge(df, df_rt, on=["ticker"], how='outer', suffixes=[None, "_eq_fun"])
    
    all_tick.append(df_merged)


df_all = pd.concat(all_tick, axis=0)
```

```python
stock_data_cols = ["ticker", "analyst_name", "dt_rating"]
df_all = df_all[df_all["dt_rating"] >= df_all["date"]].copy()
df_all.sort_values(by=stock_data_cols, inplace=True)
df_dropped = df_all.drop_duplicates(subset=stock_data_cols, keep="last").copy()
df_dropped[["ticker", "analyst_name", "dt_rating", "date", "next_day", "close", "target_price"]].head()

# At time of sale:

# Find mean of rating
# Find the average rating of the company
# Replace None target price with price at date of buy.
# build a sampler that finds the average delta for nasdaq on any given day - saves the result to a dataframe. 
# Subtract the day's delta from the stock's movement.
```

```python
df_good_tp = df_dropped

def replace_bad_target_price(row):
    result = row["target_price"]
    if result is None:
        result = row["close"] + .01
    else:
        if type(result) == str:
            result = result.replace("$", "").replace(",", "")
    return result

df_good_tp["target_price"] = df_good_tp.apply(replace_bad_target_price, axis=1).astype(float).copy()

df_good_tp.head(5)
```

```python
df_good_roi = df_good_tp

def replace_tr_rating_roi(row):
    result = row["tr_rating_roi"]
    if result is not None:
        result = result.replace("%", "").replace(",", "")
    return result

df_good_roi["tr_rating_roi"] = df_good_roi.apply(replace_tr_rating_roi, axis=1).astype(float).copy()

df_good_roi.head(5)
```

```python
df_ren = df_good_roi 

df_ren = df_ren.rename(columns={"date": "trade_date", "close": "rating_date_price", "dt_rating": "rating_date"})
df_ren.drop(columns=["open", "high", "low", "volume", "closeunadj", "lastupdated", "next_day", "rating_raw", "dividends"], inplace=True)
df_ren.reset_index(drop=True, inplace=True)

df_ren.head(5)
```
